```python
import palettable
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(nrows=6, figsize=(14, 8))
fig.subplots_adjust(bottom=0.5)

# One bit
N = 2
cmap = mpl.cm.Greys
bounds = np.linspace(0, 1, N + 1)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend="neither")
cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax1, orientation="horizontal"
)
cb.set_ticklabels([])

# Two bit
N = 4
cmap = mpl.cm.Greys
bounds = np.linspace(0, 1, N + 1)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend="neither")
cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax2, orientation="horizontal"
)
cb.set_ticklabels([])

# Three bits
N = 8
cmap = mpl.cm.Greys
bounds = np.linspace(0, 1, N + 1)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend="neither")
cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax3, orientation="horizontal"
)
cb.set_ticklabels([])

# Four bits
N = 16
cmap = mpl.cm.Greys
bounds = np.linspace(0, 1, N + 1)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend="neither")
cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax4, orientation="horizontal"
)
cb.set_ticklabels([])

# Five bits
N = 32
cmap = mpl.cm.Greys
bounds = np.linspace(0, 1, N + 1)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend="neither")
cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax5, orientation="horizontal"
)
cb.set_ticklabels([])

# Eight bits
N = 256
cmap = mpl.cm.Greys
bounds = np.linspace(0, 1, N + 1)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend="neither")
cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax6, orientation="horizontal"
)
cb.set_ticklabels([])


# Set titles
ax1.text(
    1.0,
    0.3,
    "   1 Bit - Value range: 0-1\n   (2 possible values)",
    fontdict=dict(size=18),
)
ax1.set_title("Bit depths", fontdict=dict(size=18), pad=15)
ax2.text(
    1.0,
    0.3,
    "   2 Bits - Value range: 0-3\n   (4 possible values)",
    fontdict=dict(size=18),
)
ax3.text(
    1.0,
    0.3,
    "   3 Bits - Value range: 0-7\n   (8 possible values)",
    fontdict=dict(size=18),
)
ax4.text(
    1.0,
    0.3,
    "   4 Bits - Value range: 0-15\n   (16 possible values)",
    fontdict=dict(size=18),
)
ax5.text(
    1.0,
    0.3,
    "   5 Bits - Value range: 0-31\n   (32 possible values)",
    fontdict=dict(size=18),
)
ax6.text(
    1.0,
    0.3,
    "   8 Bits - Value range: 0-255\n   (256 possible values)",
    fontdict=dict(size=18),
)

plt.gcf().subplots_adjust(right=0.45, bottom=0.05, left=0.05, top=0.95)

# Save
plt.savefig("raster_bit_depths.png", dpi=300)
```
