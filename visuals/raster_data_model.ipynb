```python
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Values
values = np.random.randint(low=0, high=20, size=(4, 4))
```

```python
# Define the subplots
fig, (ax1, ax2, cax) = plt.subplots(
    ncols=3, figsize=(20, 8), gridspec_kw={"width_ratios": [1, 1, 0.055]}
)

# Cell values
sns.heatmap(
    values,
    annot=True,
    cbar=False,
    cmap=ListedColormap(["white"]),
    ax=ax1,
    linecolor="black",
    linewidths=2,
    # Annotation style
    annot_kws=dict(size=15),
)

# Tick labels
ax1.tick_params(colors="#404040", labelsize=13)

# Colors
ax2 = sns.heatmap(
    values,
    annot=False,
    cmap="Blues",
    cbar=False,
    ax=ax2,
    linecolor="black",
    linewidths=2,
)

# colorbar
fig.colorbar(ax2.get_children()[0], cax=cax, orientation="vertical", pad=10.5)

# Arrow color
arrow_color = "black"

# Add arrows
ax1.annotate(
    "",
    xy=(0.01, -0.03),
    xytext=(0.10, -0.03),
    xycoords="axes fraction",
    arrowprops=dict(arrowstyle="<-", color=arrow_color, lw=1.5),
)
ax1.annotate(
    "",
    xy=(-0.027, 0.03),
    xytext=(-0.027, 0.10),
    xycoords="axes fraction",
    arrowprops=dict(arrowstyle="<-", color=arrow_color, lw=1.5),
)

ax2.annotate(
    "",
    xy=(0.75, -0.03),
    xytext=(1, -0.03),
    xycoords="axes fraction",
    arrowprops=dict(arrowstyle="<->", color=arrow_color, lw=2.5),
)
ax2.annotate(
    "",
    xy=(1.03, 0),
    xytext=(1.03, 0.25),
    xycoords="axes fraction",
    arrowprops=dict(arrowstyle="<->", color=arrow_color, lw=2.5),
)
cax.annotate(
    "",
    xy=(2.5, 0),
    xytext=(2.5, 0.99),
    xycoords="axes fraction",
    arrowprops=dict(arrowstyle="<->", color=arrow_color, lw=1.5, ls="--"),
)

# Add titles
fontweight = "bold"

ax1.set_title("Cell values", fontdict=dict(size=16, fontweight=fontweight), pad=10)
ax2.set_title("Value colors", fontdict=dict(size=16, fontweight=fontweight), pad=10)
cax.set_title("Legend", fontdict=dict(size=16, fontweight=fontweight), pad=10)


ax2.axes.axis("off")

# Annotation color
annot_color = "black"

# Custom texts
plt.figtext(
    0.07,
    0.0,
    "Cell position\nindex: (3, 0)",
    fontdict=dict(size=14),
    color=annot_color,
    rotation=45,
)
plt.figtext(
    0.853,
    0.17,
    "10 meters",
    fontdict=dict(size=14),
    color=annot_color,
    rotation="vertical",
)
plt.figtext(0.77, 0.06, "10 meters", fontdict=dict(size=14), color=annot_color)

plt.figtext(0.935, 0.14, "Low values", fontdict=dict(size=14), color=annot_color)
plt.figtext(0.935, 0.85, "High values", fontdict=dict(size=14), color=annot_color)


# Add arrow between the plots
plt.figtext(0.46, 0.465, "➧", fontdict=dict(size=78), color="#505050")

# Save figure
plt.savefig("raster_data_model.png", dpi=300)
```
